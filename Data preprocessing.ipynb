```python
# Schema for DF
{
    "Venue":"stadium",
    "Team-A":"a",
    "Team-A":"b",
    "Match Time": "start time",
    "Match Year": "year",
    "Umpire-1": "umpire_1",
    "Umpire-2": "umpire_2",
    "3rd Umpire": "tv_umpire",
    "Toss Winner": "toss winner",
    "Toss Decision": "Bat/Ball"
}
```

```python
# Loop for data processing

new_data = []

for items in data:
    
    venue = ((items['stadium']).split(','))[0].strip()
    team_a = items['team_a']
    team_b = items['team_b']
    #match_time = items["Hours of play (local time)"]
    year= items['Season'][0:4]
    umpire_1= items['Umpires'][0]
    umpire_2= items['Umpires'][1]
    #tv_umpire= items['TV Umpire']
    toss_winner= ((items['Toss']).split(','))[0].strip()
    toss_decision1= ((items['Toss']).split(','))[:-1]

    if 'bat' in toss_decision1: #bat=1 & field=0
        toss_decision1=1
    else:
        toss_decision1=0

    winner= items['winner']

    if winner== "NA":
        continue

    schema={
    "Venue":venue,
    "Team-A":team_a,
    "Team-B":team_b,
    #"Match Time": match_time,
    "Match Year": year,
    "Umpire-1": umpire_1,
    "Umpire-2": umpire_2,
    #"3rd Umpire": tv_umpire,
    "Toss Winner": toss_winner,
    "Toss Decision": toss_decision1,
    "Winner": winner
    }

    new_data.append(schema)

print(len(new_data))
```

```python
df = pd.DataFrame(new_data)

df.tail()
```
